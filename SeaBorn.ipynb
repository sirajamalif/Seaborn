# Line Plot

```python
var = df['age']
var_1 = df['children']
```

```python
plt.plot(var[:10],var_1[:10])  # Simple line plot using matplotlib library. Thousands of data, so did slicing
plt.show()
```

```python
sns.lineplot(x="age",y="children",data=df) # Data must be converted into DF before using seaborn
# whole dataFrame's data was used here 
```

```python
sns.lineplot(x="age",y="children",data=df[:10]) # You need to perform slicing in the whole dataFrame in seaborn
plt.show() # removes "<Axes: xlabel='age', ylabel='children'>" this line also known as ploting address 
```

```python
sns.lineplot(x="age",y="children",data=df,hue="sex",size = 50,style = 'sex',palette='Accent',markers = ['o','>'], dashes = True )
plt.grid() # adds grid lines in both axis
plt.title("Insurance")
# 'hue' adds a separate line (color) for each category in the 'sex' column
# 'size' controls the thickness or size of the line (or marker) based on the values in another column.
# 'style' parameter lets you change the lineâ€™s visual appearance (like solid, dashed, dotted, etc.) based on a categorical column.
# 'palette' controls the color scheme used for different hue levels
# 'markers' adds point markers to each data value on the line
# 'dashes' adds different dash patterns (solid, dashed, dotted, etc.) to lines based on the 'style' grouping
plt.show()
```

# Histogram Plot

```python
df = pd.read_csv(r"C:\Users\User\Downloads\insurance.csv")
df.head()
```

```python
sns.displot(df["bmi"],bins=[15,20,25,30,35,40,45,50],kde=True, rug= True,color='g',log_scale=False)
#The bins parameter defines how the data is grouped into intervals for a histogram, determining the range and number of bars shown.
# kde=True draws a smooth line that estimates the data's distribution, helping you see patterns beyond the bars.
# rug=True adds small vertical ticks (like a "rug") along the x-axis to show individual data points, giving a raw feel of where the values actually lie.
# log_scale=True transforms the axis to a logarithmic scale, making it easier to visualize data that spans several orders of magnitude (like from 1 to 10,000).
plt.show()
```

# Bar Plot

```python
df = pd.read_csv(r"C:\Users\User\Downloads\insurance.csv")
df.head()
```

```python
order_1 = ["no","yes"]
sns.barplot(x="smoker", y="bmi", hue="sex",order=order_1,hue_order=["female","male"],errorbar=('ci', 100), data=df)
# It's easy to plot multi-bar graph in seaborn with just the help of 'hue'
# The errorbar=('ci', 50) function likely specifies the error bars in a plot, where ci refers to confidence intervals and 50 sets the range or percentage (e.g., 50% confidence interval).
plt.show()
```

```python
sns.set(style="darkgrid") # Changes the background to a dark grid (gray gridlines over a light gray background)
sns.barplot(x="age", y="bmi",orient='h',color='y',palette='icefire', data=df[:20])
# when doing orient='h' make sure x and y both consist of numerical values
#plt.ylim(15, 45) # Adjust the y-axis range here
plt.show()
```
